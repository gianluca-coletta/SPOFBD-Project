```python
# Import libraries and modules for data analysis 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the data

df = pd.read_csv('../Dataset/dataset.csv')
```

```python
# Display the first 5 rows of the dataframe

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Count the number of missing values in each column

df.isnull().sum()   
```

```python
# Change the column names quality: if value = bad then 0 else 1

df['quality'] = df['quality'].apply(lambda x: -1 if x == 'bad' else 1)
```

```python
# Display the first 5 rows of the dataframe

df.head()
```

```python
# Bar chart
ax = df['quality'].value_counts().plot(kind='bar', rot=0)  # La parte chiave è rot=0

# Set the labels for the x and y axes and the title
plt.xlabel("Good or Bad")
plt.ylabel("Count")
plt.title("Quality Wine")

# Show the chart
plt.show()
```

```python
# Correlation matrix heatmap 

# Set the width and height of the figure

plt.figure(figsize=(10,6))

# Generate correlation matrix

corr = df.corr()

# Plot heatmap of annotated correlations

sns.heatmap(corr, annot=True, cmap='coolwarm')

# Show the plot

plt.show()
```

```python
# Move the 'quality' column to the front of the dataframe 

df = df[['quality', 'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']] 
```

```python
# Create a new dataset and save it into csv file

df.to_csv('../dataset/dataset_clean.csv', index=False)
```
